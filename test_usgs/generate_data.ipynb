```python
import csv
import os

# Add Ferrosilicon to the list of new materials
new_materials = [
    # ... (previous materials remain unchanged)
    ("Selenium", "{\"Se\": 1.0}"),
    # Add Ferrosilicon here
    ("Ferrosilicon", "{\"Si\": 0.75, \"Fe\": 0.25}")
]

def update_csv(file_path, new_data):
    try:
        # Check if file exists
        if not os.path.exists(file_path):
            raise FileNotFoundError(f"The file {file_path} does not exist.")

        # Append new materials to the existing CSV
        with open(file_path, 'a', newline='') as f:
            writer = csv.writer(f)
            for material, composition in new_data:
                writer.writerow([material, composition])

        print(f"CSV file {file_path} updated successfully.")
    except FileNotFoundError as e:
        print(f"Error: {e}")
    except PermissionError:
        print(f"Error: Permission denied. Unable to write to {file_path}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Update the CSV file
csv_filename = 'sub_materials_database.csv\sub_materials_database_sub_materials_database.csv'

update_csv(csv_filename, new_materials)
```
